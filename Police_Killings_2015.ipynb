```python
df_kills = pd.read_csv('killed_by_police.csv',delimiter=',',low_memory=False, encoding = 'latin-1')
States_label = pd.read_csv('states.csv',delimiter=',',low_memory=False, index_col=0)


#df_kills.to_csv('mapping.csv')
#df_short = pd.read_csv('mapping.csv', usecols=['person', 'age', 'gender','raceethnicity','city','state','pop','p_income','urate'])
#df_short.columns = ['person', 'age', 'gender','raceethnicity','city','state','pop','p_income','urate']

df_kills['state'] = df_kills['state'].map(States_label['State'])
df_kills.to_csv('killed_by_police_Noabr.csv')
df_kills
```

1. Changing the Abbreviations of the state column to names now allows me to create a maplot in tableau which will allow me to check if the more populated the state, then the more the killings by the police
2. To do that we create a much simpler data set 
   by having only the necessary columns: "person, state, pop: tract population"
3. Then groupby the table sum the tract population of each state that we have based on our data
   and finally count the people died in 2015 in each state

```python
df_kills.to_csv('kp_map.csv')

df_psp = pd.read_csv('kp_map.csv',delimiter=',',low_memory=False, encoding = 'latin-1',usecols=['person', 'state','pop'])
df_psp.columns = ['person', 'state','pop']


df_psp_final = df_psp.groupby('state').agg({'pop':['sum'],'person':['count']})

df_psp_final = df_psp_final.sort_values([('person','count')], ascending=False)
df_psp_final
```

```python
#Calculating the correlation of person with the tract census population of states
#a high positive correlation A positive indicates that if one variable increases, the other increases also. 

df_psp_final['pop','sum'].corr(df_psp_final['person','count'])
```

```python
#Remove sum and count headers name because tableau was reading sum and count as values of the first row
df_psp_final.columns = df_psp_final.columns.get_level_values(0)
df_psp_final
```

```python
df_psp_final.to_csv('df_psp_corr.csv')
```

## Killings by Age grouping

```python
#We create a table based on age. Recode data by creating groups of age 
#To do so age column type needs to be numeric, therefore we change the Unknown word in the column with the pandas NaN

df_ageT = pd.read_csv('kp_map.csv',delimiter=',',low_memory=False, encoding = 'latin-1',usecols=['person','age','gender','raceethnicity','month','cause','armed','p_income','h_income','pov','urate'])
df_ageT.columns = ['person','age','gender','raceethnicity','month','cause','armed','p_income','h_income','pov','urate']
df_ageT['age'] = df_ageT.age.replace('Unknown',np.NaN)


df_ageT['age'] = df_ageT.age.astype('float')

def age_groups(series):
    if series < 18:
        return "-18"
    elif 18 <= series < 25:
        return "18-24"
    elif 25 <= series < 35:
        return "25-34"
    elif 35 <= series < 45:
        return "35-44"
    elif 45 <= series < 55:
        return "45-54"
    elif 55 <= series < 65:
        return "55-64"
    elif 65 <= series < 75:
        return "65-74"
    elif 75 <= series:
        return "74+"

df_ageT['age'] = df_ageT['age'].apply(age_groups)

df_ageT['age'].value_counts(sort=True)


```

```python
df_kills['age'] = df_kills.age.replace('Unknown',np.NaN)
##
df_kills['age'] = df_kills.age.astype('float')
less_18 = df_kills[(df_kills.age < 18)]
more_74 = df_kills[(df_kills.age > 74)]
#more_74
#less_18
```

### Check the relationship between person armed with firearm and cause of death by gunshot 

```python
#Creating a table with dummy variables to calculate later on the relationship between the cause of death (gunshot)
#and armed person

df_armNgun = pd.read_csv('kp_map.csv',delimiter=',',low_memory=False, encoding = 'latin-1',usecols=['armed','cause'])
df_armNgun.columns = ['cause','armed']

dum_cause = pd.get_dummies(df_armNgun['cause'])
del dum_cause[dum_cause.columns[-1]]

##############################################
dum_arm = pd.get_dummies(df_armNgun['armed'])

del dum_arm[dum_arm.columns[-1]]

new_armNgun = pd.concat([dum_cause, dum_arm], axis=1)

#del new_armed['armed']
new_armNgun
```

```python
#Check if there is a correlation (relationship) between the person who died by gunshot and if they were armed with
# a firearm weapon
correlation_armNgun = new_armNgun['Gunshot'].corr(new_armNgun['Firearm'])
correlation_armNgun
```

Based on the -0.04508 correlation found, the person died by gunshot do not affect the person armed with a firearm, so there is no relationship between the two.
